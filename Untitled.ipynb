```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
tt=pd.read_csv("./train.csv")
```

```python
print(tt);
a=tt.isnull().sum()
print("Percent of missing data",'\n',round(a/len(tt),2)*100)
```

```python
ts=tt.iloc[:,[2,3,4,5,6,7,8,9,11]]#training set
tsy=tt.iloc[:,1]#training set y
```

```python
plt.hist(ts["Age"],25,density=True);
plt.show();
print(ts["Age"].describe(),'\n','-------------------------------')
print(ts["Age"].sort_values())
```

```python
fig, axs=plt.subplots(3,3);
fig.set_size_inches(20, 25);
fig.tight_layout();
axs[0,0].hist(ts["Age"][ts["Pclass"]==1]);
axs[0,0].set_title('Age in first class');
axs[0,1].hist(ts["Age"][ts["Pclass"]==2]);
axs[0,1].set_title('Age in second class');
axs[0,2].hist(ts["Age"][ts["Pclass"]==3]);
axs[0,2].set_title('Age in third class');
axs[1,0].hist(ts["Age"][ts["Sex"]=='male']);
axs[1,0].set_title("No males");
axs[1,1].hist(ts["Age"][ts["Sex"]=='female']);
axs[1,1].set_title("No females");
axs[1,2].scatter(ts["Fare"][ts["Fare"]<300],ts["Age"][ts["Fare"]<300]);
axs[1,2].set(xlabel="fare",ylabel="age",title="relation x-fare y-age");
plt.show();
```
